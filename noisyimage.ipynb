```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Path to your image (you can replace this with the actual path)
image_path = "/kaggle/input/cat-image/noisy_image.png"

# Load the noisy image
original_image = cv2.imread(image_path)
noisy_image = original_image.copy()

# Loop to apply median filtering multiple times
for i in range(50):
    # Apply Median Filtering to remove salt-and-pepper noise
    denoised_image = cv2.medianBlur(noisy_image, 5)
    
    # Set the current denoised image as the noisy image for the next iteration
    noisy_image = denoised_image  # Use the current denoised image as input for the next iteration.

# Display the original and final denoised images
plt.figure(figsize=(10, 5))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis("off")

# Final Denoised Image
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(denoised_image, cv2.COLOR_BGR2RGB))
plt.title("Final Denoised Image")
plt.axis("off")

plt.show()
```
